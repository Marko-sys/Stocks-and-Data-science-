```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression 

csv_path='revenue.csv'
df=pd.read_csv(csv_path, nrows=10)
df.head()
df_new=df.iloc[:, [1,2,3,5]]
df_new
print(df_new)
print("-----------------------------------------------------------------")


usa= len(df_new[df_new['country'] == 'United States'])
chn=len(df_new[df_new['country'] == 'China'])
sa=len(df_new[df_new['country'] == 'Saudi Arabia'])

print("USA:", usa)
print("China:",chn)
print("Saudi:",sa)

y=np.array([usa,chn,sa])
my_labels=["USA","China","Saudi" ]
my_colors=["Blue","Red","Green"]
plt.pie(y,labels=my_labels, autopct='%.1f%%',wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'}, 
textprops={'size': 'x-large', 'color':'white'}, shadow=True, colors=my_colors)
plt.legend(title="country")
plt.show()

print("-----------------------------------------------------------------")

rev=df_new['revenue (Billion USD)']
nam=df_new["" 'Symbol']

plt.barh(nam,rev, height=0.6 )
for index, value in enumerate(rev):
    plt.text(value, index,
             str(value))
plt.title("Top 10 Companies in the world, Revenue wise - 02-07-2022")
plt.ylabel("Company Stock Symbol")
plt.xlabel("Reveneu in Billion ($)")
plt.show()

```
